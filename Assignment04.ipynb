```python
import geemap,geemap as import_map
```

```python
import ee
```

```python
ee.Authenticate()
```

```python
ee.Initialize(project = 'rohitshinde4680')
import os
```

```python
m=import_map.Map()
```

```python
India_Districts = ee.FeatureCollection('projects/rohitshinde4680/assets/India_districts')
m.addLayer(India_Districts,{},'India_Districts') #dislpaying the shapefile in the interactive map
```

```python
treeCover = ee.ImageCollection("MODIS/006/MOD44B").filterDate("2016-01-01","2016-12-31").select("Percent_Tree_Cover").mean().clip(India_Districts)
Visualize_Tree = {"min":0, "max":100, 'palette': ['bbe029', '0a9501', '074b03']}
m.addLayer(treeCover, Visualize_Tree, 'treeCover')
m
```

```python
output_path=os.path.join(os.path.expanduser("~"), "Downloads")
out_dsr_stats = os.path.join(output_path, "Treecover_2016.csv")
if not os.path.exists(output_path):
    os.makedirs(output_path)
import_map.zonal_stats(treeCover,India_Districts, out_dsr_stats, stat_type="MEAN", scale=250) #download csv for dsr 2016
```
